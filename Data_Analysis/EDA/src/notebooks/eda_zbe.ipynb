Función para eliminar los valores erroneos de los datos

```python
def limpiar_valores_erroneos(df: pd.DataFrame) -> pd.DataFrame:
    df = df.reset_index(drop=True)

    for i, row in df.iterrows():
        for valor in range(len(row)):
            if 'N' == row.iloc[valor]:
                if valor < 10:
                    if i > 0:
                        fila_anterior = df.loc[i-1]
                        row.iloc[valor - 1] = fila_anterior.iloc[-2]
                        row.iloc[valor] = 'V'
                    else:
                        row.iloc[valor - 1] = 0
                        row.iloc[valor] = 'V'
                else:
                    row.iloc[valor - 1] = row.iloc[valor - 3]
                    row.iloc[valor] = 'V'

        df.loc[i] = row
        
    return df
```

```python
def limpiar_columnas_validador(df: pd.DataFrame) -> pd.DataFrame:
    for titulos in df:
        if titulos.startswith('V'):
            df = df.drop(titulos, axis=1)

    return df
```

Ahora voy a guardar todos los meses en una sola fila. Y luego ir metiendo cada años en cada fila

```python
def pasar_datos_a_fila(df: pd.DataFrame) -> pd.DataFrame:
    anio = df['ANO'].drop_duplicates()
    for d in anio:
        anio = d
    df = df.drop(["MAGNITUD", "ANO"], axis=1)
    if (len(df) > 0):
        fila1 = df.iloc[0]
        fila2 = df.iloc[1]
        fila3 = df.iloc[2]
        fila4 = df.iloc[3]
        fila5 = df.iloc[4]
        fila6 = df.iloc[5]
        fila7 = df.iloc[6]
        fila8 = df.iloc[7]
        fila9 = df.iloc[8]
        fila10 = df.iloc[9]
        fila11 = df.iloc[10]
        fila12 = df.iloc[11]
        nuevo_df = pd.concat([fila1, fila2, fila3, fila4, fila5, fila6, fila7, fila8, fila9, fila10, fila11, fila12]).to_frame().T.reset_index(drop=True)
        nuevo_df = nuevo_df.drop("MES", axis=1)

        nuevo_df.columns = range(1, len(nuevo_df.columns) + 1)
        nuevo_df['ANIO'] = anio
        nuevo_df = nuevo_df.set_index('ANIO')

        return nuevo_df.T
    else:
        return df    
```

Función para crear las trazas de los gráficos

```python
def crear_trazas(fila: pd.DataFrame, anio: int, color: str) -> go.Scatter:
    try:
        traza = go.Scatter(
                    x = fila.index,
                    y = fila[anio],
                    name = str(anio),
                    mode = 'lines',
                    marker = dict(color = color))
    except:
        traza = go.Scatter(
                    x = np.zeros(372),
                    y = np.zeros(372),
                    name = str(anio),
                    mode = 'lines',
                    marker = dict(color = color))
        
    return traza
```

Funcion para crear el gráfico

```python
colores = ['rgba(0, 255, 0, 0.8)', 'rgba(255, 0, 0, 0.8)', 'rgba(0, 0, 255, 0.8)', 'rgba(180, 26, 200, 0.8)', 'rgba(80, 80, 200, 0.8)', 'rgba(80, 200, 80, 0.8)', 'rgba(200, 80, 80, 0.8)']
anios = [2016, 2017, 2018, 2019, 2020, 2021, 2022]

def crear_grafico(df: pd.DataFrame, titulo: str) -> go.Figure:
    trazas = []

    for index, df in enumerate(df):
        trazas.append(crear_trazas(df, anios[index], colores[index]))

    layout = dict(title = titulo, xaxis= dict(title= 'Día del año',ticklen= 5))
    fig = go.Figure(data = trazas, layout=layout)

    return fig
```
