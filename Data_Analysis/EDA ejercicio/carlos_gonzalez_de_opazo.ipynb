Vemos que hay datos que no están colocados en sus columnas por lo que vamos a proceder a actualizar esos valores, algunos son terminales básicos que llevan un sistema operativo propietario que no es Android, iOS, blackberryOS, symbian, etc. Por lo que voy a sustituir estos valores nulos por Propietary

```python
df["Bluetooth"] = df["card"].str.find("Bluetooth") != -1
df["os"] = df["camera"].str.find("Android") != -1

# Buscar el valor "Android" en la columna "MARCA"
valor = df["camera"].str.contains("Android")

# Si el valor existe, asignarlo a la columna "OS"
df["os"] = df["camera"].where(valor, "No existe")

df.loc[df["os"].isnull()]
```

```python
df["os"].replace(np.nan, "Propietary", inplace=True)
```

Me voy a fijar en los tipos de conectividad móvil (2g, 3g, 4g, 5g, bluetooth y wifi) También en la memoria RAM, en si tienen o no cámara, tarjeta de memoria, si el sistema operativo es Android o iOS y el precio

Primero convierto el precio de rupias a Euros

```python
def rupias_a_euros(rupias: float):
    euros = rupias * 0.011
    return euros

df
```

```python
# Eliminamos el simbolo y la coma por un . para poder hacer la conversión de la columna a float

df['price'] = df['price'].str.replace("₹", "")
df['price'] = df['price'].str.replace(",", "")
df['price'] = df['price'].astype(float)

```

```python
df['price'] = df['price'].apply(lambda x: rupias_a_euros(x))

df
```
